```python
lines = cv2.HoughLines(Simple_Canny, 1, np.pi * 2 / 180, 170)
print(lines.shape)

def DrawLines(img, lines):
    # Draw the detected lines on the original image
    img = np.copy(img)
    for line in lines:
        rho, theta = line[0]
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * a)
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * a)
        cv2.line(img, (x1, y1), (x2, y2), (255, 255, 255), thickness=2)
        
    return img

img = DrawLines(Simple_Canny, lines)
ImagePlotter(img).plot_image('Simple_Canny_with_Lines')
```

```python
def split_image_edged_nxn_sections(image, n):
    
    # Get the size of the image
    height, width = image.shape[:2]
    
    img = np.copy(image)

    if img.ndim==2:
        img = np.expand_dims(img, axis=2)
        
    img = ColorSpace(img).rgb_to_grayscale()
    img = img.astype(np.uint8)
    
    # Turn the image into edge
    img = cv2.Canny(img, 50, 150, apertureSize=3)
    
    ImagePlotter(img).plot_image('Simple_Canny_with_Lines')

    # Calculate the height of each section
    section_height = int(np.ceil(height / n))

    # Calculate the width of each section
    section_width = int(np.ceil(width / n))

    # Initialize the list to store the sections
    image_list = []
    to_origin = []
    for i in range(0, height, section_height):
        for j in range(0, width, section_width):
            section = np.array(img[i:i + section_height,j:j + section_width]).astype(np.uint8)
            x = j * section_width
            y = i * section_height
            to_origin.append((x, y))
            image_list.append(section)
    
    return image_list, to_origin
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from numba import jit
import numba as nb

def Get_Tiled_Hough_Intercepts(image_list, to_origin):
    
    @jit(nopython=True)
    def process_line(line, image_section, origin):
        # Calculating the xy coordinates of the lines' intersection with the image section borders
        height, width = image_section.shape[:2]

        rho, theta = line[0]
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = (x0 + width * (-b)).astype(np.uint8) 
        y1 = (y0 + height * (a)).astype(np.uint8)
        x2 = (x0 + width * b).astype(np.uint8)
        y2 = (y0 - height * (a)).astype(np.uint8)
        
        # Return start (x, y) , end (x, y). Origin is used to find coordinates in the global space
        return ((x1 + origin[0], y1 + origin[1]), (x2 + origin[0], y2 + origin[1]))
    
    lines = []
    for section, origin in zip(image_list, to_origin):
        
        print("Section shape:", section.shape)
        print("Section type:", section.dtype)
        
        h_lines = None
        
        # Getting edges and lines from the image section
        h_lines = cv2.HoughLines(section, 1, np.pi / 180, 50)
        
        # We convert every line into xy endpoints that intercept
        # the edges of each window.
        if h_lines is not None:
            for line in range(len(h_lines)):
                lines.append(process_line(h_lines[line], section, origin))
            
    return lines
    
```

```python
# This function takes all the lines and returns lines that meet particular criteria.
def aggregate_lines(lines, n, distance_threshold=5, line_depth_threshold=5):
    linear_plots = list(lines)

    @jit(nopython=True)
    def distance(point1, point2):
        x1, y1 = point1[0], point1[1]
        x2, y2 = point2[0], point2[1]
        return ((x1 - x2)**2 + (y1 - y2)**2)**0.5
    
    def aggregate(linear_plots):
        # Aggregates linear_plots by reference
        i = 0

        # Since the length of linear_plots is dynamic, we need to re-evaluate it's length
        # every iteration.
        while i < len(linear_plots):
            existing_line = linear_plots[i]
            j = 0
            while j < len(linear_plots):
                potential_line = linear_plots[j]

                if i == j:
                    j += 1
                    continue

                # If the potential line is too far (x), continue
                if potential_line[1][0] > existing_line[-1][0] + distance_threshold:
                    j += 1
                    continue

                # If the potential line is too far (y), continue
                if potential_line[1][1] > existing_line[-1][1] + distance_threshold:
                    j += 1
                    continue

                # If potential line is overlapping, delete and continue
                if distance(existing_line[0], potential_line[0]) < distance_threshold:

                    # Delete potential line
                    del linear_plots[j]
                    j += 1
                    continue

                # if distance between end of existing line and start of potential line
                # is less than threshold, apend the end to the end of existing line
                # and delete the potential line
                if distance(existing_line[-1], potential_line[0]) < distance_threshold:
                    existing_line.append(potential_line[1])
                    del linear_plots[j]
                    j += 1
                    continue

                # If none of these conditions are triggered, then pass to the next
                j += 1

    # Run through this n times, each run through builds a line for one section
    # Where an image is segmented into n sections, the possible line will be
    # Approximately n deep.
    for count in range(0, n):
        aggregate(linear_plots)
    
    # Filter the results by length
    for i, line in enumerate(linear_plots):
        if len(line) < line_depth_threshold:
            del linear_plots[i]

    # Each line is a series of coordinates that a regression can be run on.
    return linear_plots
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt


def plot_polyfit_on_image(image, lines, degree=2):
    for line in lines:
        x = [point[0] for point in line]
        y = [point[1] for point in line]

        # Fit polynomial curve to line
        coeffs = np.polyfit(x, y, degree)
        poly_func = np.poly1d(coeffs)

        # Generate points for plotting curve
        fit_x = np.linspace(min(x), max(x), 100)
        fit_y = poly_func(fit_x)

        # Plot regression curve on image
        plt.plot(fit_x, fit_y, 'r-')

    # Show image with regression curves
    plt.imshow(image)
    plt.show()
```

```python
def Linear_Plot_Pipeline(image, sections, aggregation_depth, line_depth, plot_degree = 2):
    image_dict, to_origin = split_image_edged_nxn_sections(image, sections)
    line_segments = Get_Tiled_Hough_Intercepts(image_dict, to_origin)
    lines = aggregate_lines(line_segments, aggregation_depth, line_depth)
    plot_polyfit_on_image(image, lines, plot_degree)
    return lines
```

```python
from numba import jit

# @staticmethod
@jit(nopython=True)
def min(sub_image, weight_matrix):
    # Create a flat version of the input array
    flat_arr = sub_image.flatten()
    # Create a flat version of the weight matrix
    flat_weight_matrix = np.array(weight_matrix).flatten()

    nonzero_mask = flat_weight_matrix != 0
    # return only non-zero values
    nonzero_pixels = flat_arr[nonzero_mask]

    return np.min(nonzero_pixels)
```
