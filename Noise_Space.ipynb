```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load an image using OpenCV
orange1 = cv2.imread('orange1.jpg')

# Convert the image to RGB and a NumPy ndarray
rgb_image = cv2.cvtColor(orange1, cv2.COLOR_BGR2RGB)
orange1 = np.array(rgb_image)
```

```python
import noisegenerators as ng
import imageplotter as ip


ip.ImagePlotter(ng.gaussian_noise(orange1, 15, 40)).noise_with_histogram(title = 'Gaussian')
```

```python
noisy_image = ng.add_gaussian_noise(orange1, 30, 25)


ip.ImagePlotter(noisy_image).image_with_histogram(title='Noisy Gaussian')
```
